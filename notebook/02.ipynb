```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700])
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000, 319000, 255000]) 
```

```python
w = 0.1 
b = 0.1 
```

```python
def hypothesis (x):
    return w * x + b
```

```python
def cost_fucn (X , y):
    m = len (y)
    prediction = hypothesis (X)
    return (1/(2*m))* np.sum(((prediction - y) **2))
```

```python
def GD (X, y, w, b, lr, epoch):
    m = len (y)
    cost_history = []
    for _ in range (epoch):
        prediction = hypothesis (X)
        dw = (1 / m) * np.sum((prediction - y) * X)
        db = (1 / m)* np.sum((prediction - y))
        w -= lr * dw
        b -= lr * db
        cost = cost_fucn(X, y)
        cost_history.append(cost)

    return w, b, cost_history
```

```python
lr = 0.00001
epoch = 1000
w, b, cost_history = GD (X, y, w, b, lr, epoch)
```

```python
new_house_size = 2000
new_house_price = hypothesis(new_house_size)
print (f"the price of the house with {new_house_size} is equal to the {new_house_price:.2f}")
```

```python
plt.scatter(X, y, label= 'Actual Data' , color = 'blue')
plt.xlabel ("House size in SQR feet")
plt.ylabel ("house price")
plt.title ("scatter plot of the house size vs price")
plt.legend()
plt.show ()
```

```python
plt.scatter(X, y, label='Actual Data', color='blue')
plt.plot(X, hypothesis(X), label='Regression Line', color='red')
plt.xlabel('House Size (sq. ft)')
plt.ylabel('House Price')
plt.title('Regression Line and Actual Data')
plt.legend()
plt.show()
```

```python
plt.plot(range(epoch), cost_history)
plt.xlabel('Iteration')
plt.ylabel('cost history')
plt.title ('cost fun vs iteration')
plt.show()
```
