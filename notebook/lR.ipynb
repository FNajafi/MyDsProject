```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1400, 3],
    [1600, 3],
    [1700, 2],
    [1875, 4],
    [1100, 2],
    [1550, 3],
    [2350, 4],
    [2450, 5],
    [1425, 3],
    [1700, 2]
])
```

```python
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000, 319000, 255000])
```

```python
w1 = 0.1
w2 = 0.1
b = 0.1
```

```python
def hypothesis (x):
    return w1 * x[0] + w2 * x[1] + b 
```

```python
def cost_fucn (X, y):
    m = len (y)
    prediction = np.array ([hypothesis(x) for x in X])
    return (1/(2*m))* np.sum(((prediction - y) **2))
```

```python
def GD (X, y, w1, w2, b, lr, epoch):
    m = len (y)
    cost_history = []
    for i in range (epoch):
        prediction = np.array ([hypothesis(x)for x in X])
        dw1 = (1/ m)* np.sum((prediction - y) * X[:,0])
        dw2 = (1/ m)* np.sum((prediction - y) * X[:,1])
        db = (1/ m)* np.sum((prediction - y))
        w1 -= lr*dw1
        w2 -= lr*dw2
        b -= lr* db
        cost_value = cost_fucn (X, y)
        cost_history.append(cost_value)
    return w1 , w2, b, cost_history
```

```python
lr = 0.00001
epoch = 1000
w1, w2, b, cost_history = GD (X, y, w1, w2, b, lr, epoch)
```

```python
new_house_size = 2000
n_bedrooms = 4
predicted_price = hypothesis([new_house_size, n_bedrooms])
print(f"Predicted price for a house with size {new_house_size} sq. ft and {n_bedrooms} bedrooms: ${predicted_price:.2f}")
```
